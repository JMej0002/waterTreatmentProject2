```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('../Calculations/MWC_ETP_Daily_InfluentQuality_From2014_698890028305835467.csv')

# Drop NaN values
bod_values = df["BOD_mg.L-1"].dropna()

# Compute mean
avg_BOD = bod_values.mean()

# Plot
fig, ax = plt.subplots()
ax.boxplot(bod_values, vert=False, showmeans=True, patch_artist=True)

# Title and labels
ax.set_title('Box Plot of BOD_mg.L-1')
ax.set_xlabel('BOD (mg/L)')

# Add text label slightly above and to the right of the mean marker
ax.text(avg_BOD + 1, 1.1, f"Mean: {avg_BOD:.2f}", ha='left', va='center', fontsize=10, color='blue')

output_path='../images/boxPlot_BOD.png'
plt.savefig(output_path, dpi=300)

plt.tight_layout()
plt.show()



```
